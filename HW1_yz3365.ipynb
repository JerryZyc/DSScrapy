```python
btrain.corr()['medv']
# 'rm' and 'lstat' are most correlated with housing price.
```

```python
rm_std = (btrain.rm-btrain.rm.min())/(btrain.rm.max()-btrain.rm.min())
age_std = (btrain.age-btrain.age.min())/(btrain.age.max()-btrain.age.min())

new_feature=pd.DataFrame({'ID':btrain['ID'],'house_condition':(0.69*rm_std-0.36*age_std)})
medv=pd.DataFrame({'ID':btrain['ID'],'medv':(btrain['medv'])})

house_condition=pd.merge(new_feature,medv)
print(house_condition.head())
house_condition.plot('house_condition','medv',kind='scatter')
#The new feature is a combination of 'rm' and 'age', regarding as house's condition.
```

```python
feature_cols = ['rm','lstat','crim','zn','indus','nox','age','dis','rad','tax','ptratio','black']
X = btrain.loc[:, feature_cols]
X.shape
```

```python
y=btrain.medv
y.shape
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()

logreg.fit(X, y.astype('int'))
```

```python
btest = pd.read_csv('test.csv')
btest.head()
```

```python
X_new = btest.loc[:, feature_cols]
X_new.shape
new_medv = logreg.predict(X_new)
medv_test=pd.DataFrame({'ID':btest['ID'],'medv':(new_medv)})
medv_test.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(btest['ID'],new_medv)
```
